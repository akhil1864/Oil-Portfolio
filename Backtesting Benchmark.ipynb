```python
import yfinance as yf
import pandas as pd
import numpy as np
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt

benchmark_tickers = ['XOM','CVX','SHEL','TTE','COP','ENB','BP','PBR','WMB','EPD',
'EOG','CNQ','EQNR','ET','KMI','E','TRP','PSX','LNG','MPC',
'MPLX','SU','OKE','VLO','OXY','IMO','FANG','TRGP','CVE','CQP',
'DVN','PBA','EC','CTRA','WES','VNOM','YPF','PAA',
'PR','OVV','DTM','APA','AR','DINO','HESM','AM','RRC','SUN',
'KNTK','CHRD','MTDR','MGY','GLNG','CRK','PAGP','TGS','VIST',
'CRC','CNX','UGP','MUR','SM','CIVI','GPOR','PBF','EE','CVI',
'BSM','CRGY','NOG','DKL','GEL','CSAN','BTE','TALO','MNR','PARR',
'KRP','DMLP','VET','VTS','HPK','KOS','INR','TXO','GRNT','REPX',
'VTLE','SD','EGY','OBE','GPRK','GFR','HUSA','ANNA','SJT','WTI',
'PNRG','BRY']

# --------- HISTORIC RANGE (Jan 1st 2019 → today) ----------
start = "2019-01-01"
end = pd.Timestamp.today().normalize()

# --------- DOWNLOAD PRICE & VOLUME DATA ----------
raw = yf.download(benchmark_tickers, start=start, end=end,
                  auto_adjust=False, group_by='column', progress=False)

prices = raw['Adj Close'].copy()
volumes = raw['Volume'].copy()

# Drop tickers with no data
keep = prices.columns[prices.notna().any()]
prices = prices[keep]
volumes = volumes[keep]

# --------- DAILY RETURNS ----------
ret = prices.pct_change()

# --------- 30D AVG VOLUME WEIGHTS (no look-ahead) ----------
avg_vol_30 = volumes.rolling(window=30, min_periods=15).mean()
vol_sum = avg_vol_30.sum(axis=1).replace(0, np.nan)
volw = avg_vol_30.div(vol_sum, axis=0)

# shift weights by 1 day to avoid look-ahead bias
volw_shift = volw.shift(1)

# --------- ALIGN / CLEAN ----------
ret, volw_shift = ret.align(volw_shift, join='inner', axis=0)
ret = ret.dropna(how='all')

# fill missing weights with 0 and renormalize
W = volw_shift.fillna(0.0)
row_sums = W.sum(axis=1).replace(0, np.nan)
W = W.div(row_sums, axis=0).fillna(0.0)

# --------- BENCHMARK SERIES ----------
benchmark_ret = (ret * W).sum(axis=1)
benchmark_index = (1 + benchmark_ret).cumprod() * 100

# --------- RISK-FREE DATA (3M T-BILL from FRED) ----------
rf = pdr.DataReader('DGS3MO', 'fred', start, end) / 100.0
rf = rf.ffill()
rf_daily = rf['DGS3MO'] / 252.0
rf_daily = rf_daily.reindex(benchmark_ret.index).ffill()

# --------- EXCESS RETURNS ----------
excess_ret = benchmark_ret - rf_daily

# --------- SUMMARY WITH PROPER SHARPE ----------
n = excess_ret.shape[0]
ann_return = (1 + excess_ret).prod()**(252/n) - 1
ann_vol = excess_ret.std() * np.sqrt(252)
sharpe = ann_return / ann_vol if ann_vol != 0 else np.nan

print("Start:", benchmark_index.index.min().date(), " End:", benchmark_index.index.max().date())
print("Trading days:", n)
print("Annualized Excess Return:", round(ann_return, 4))
print("Annualized Vol:", round(ann_vol, 4))
print("Sharpe (rf = 3M T-bill):", round(sharpe, 3))

# --------- ROLLING SHARPE ----------
ROLLING_WINDOW = 252
rolling_sharpe_bench = excess_ret.rolling(ROLLING_WINDOW).apply(lambda x: np.sqrt(252) * x.mean() / x.std(), raw=True)


plt.figure(figsize=(9,5))
plt.plot(rolling_sharpe_bench, label=f"Benchmark (window={ROLLING_WINDOW}d)")
plt.axhline(0, linewidth=1)
plt.title("Sharpe Ratio (annualized, rf = 3M T-bill)")
plt.ylabel("Sharpe")
plt.xlabel("Date")
plt.legend()
plt.tight_layout()
plt.show()
```
